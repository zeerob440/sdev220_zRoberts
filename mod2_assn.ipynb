```python
# 6.2
# appears to be a number guessing game for the computer to play. 
guess_me: int = 7
number: int = 1
too_low: str = 'too_low'
found_it: str ='found it!'
this_is_the_end: str = 'End of Program'
strong_sentinel_flag: str = 'oops'



while number <= guess_me:
    if number < guess_me:
        print(too_low)
    if number == guess_me:
        print(found_it, f'number is {number}!')  
    if number > guess_me:
        print(strong_sentinel_flag)
        break
    number += 1

print(this_is_the_end)




```
