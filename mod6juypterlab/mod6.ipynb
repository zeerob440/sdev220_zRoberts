```python
# 13.1
from datetime import datetime 

mil_date: str = datetime.today().strftime("%d-%b-%Y").upper()

with open("today.txt", "w") as file:
    file.write(mil_date)

# print check
# print(mil_date)
# correct string saved in today.txt
```
