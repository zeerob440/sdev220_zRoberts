```python
#13.3 

from datetime import datetime

# parse the today_string

parsed_today_string: str = datetime.strptime(today_string, "%d-%b-%Y")

most_logical_format: str = "The most logical date format increments from smallest unit of time to largest, like numbers."
parsed_today_string_day: str = f" The day is: {parsed_today_string.day}."
parsed_today_string_month: str = f" The month is: {parsed_today_string.month}."
parsed_today_string_year: str = f" The year is: {parsed_today_string.year}."

print(most_logical_format + parsed_today_string_day + parsed_today_string_month + parsed_today_string_year)
```
